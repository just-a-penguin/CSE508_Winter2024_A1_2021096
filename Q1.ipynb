```python
import os
import string
from collections import defaultdict

def tokenize(text):
    """
    Tokenizes the input text into individual words.
    """
    # Remove punctuation and convert to lowercase
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator).lower()
    # Tokenize by whitespace
    tokens = text.split()
    return tokens

def create_inverted_index(folder_path):
    """
    Creates a unigram inverted index from a folder containing multiple files.
    """
    inverted_index = defaultdict(list)  # Dictionary to store inverted index

    # Iterate through all files in the folder
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            with open(file_path, 'r', encoding='utf-8') as file:
                text = file.read()
                tokens = tokenize(text)
                # Create the inverted index for the current document
                doc_id = int(''.join(filter(str.isdigit, filename)))  # Extract document ID from filename
                for token in set(tokens):  # Use set to remove duplicate tokens
                    inverted_index[token].append(doc_id)

    return inverted_index


folder_path = 'text_files/'
inverted_index = create_inverted_index(folder_path)

# Print the inverted index
for term, doc_ids in inverted_index.items():
    print(f"{term}: {doc_ids}")

vocab=set([i for i in range(1,1000)])
```

```python
import os
import string
from collections import defaultdict

def tokenize(text):
    """
    Tokenizes the input text into individual words.
    """
    # Remove punctuation and convert to lowercase
    translator = str.maketrans('', '', string.punctuation)
    text = text.translate(translator).lower()
    # Tokenize by whitespace
    tokens = text.split()
    return tokens

def create_positional_index(folder_path):
    """
    Creates a positional index from a folder containing multiple files.
    """
    positional_index = defaultdict(lambda: defaultdict(list))  # Dictionary to store positional index

    # Iterate through all files in the folder
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            with open(file_path, 'r', encoding='utf-8') as file:
                text = file.read()
                tokens = tokenize(text)
                # Create the positional index for the current document
                doc_id = int(''.join(filter(str.isdigit, filename)))  # Extract document ID from filename
                for position, token in enumerate(tokens):
                    positional_index[token][doc_id].append(position)

    return positional_index

folder_path = 'text_files/'
positional_index = create_positional_index(folder_path)

# Print the positional index
for term, postings in positional_index.items():
    print(f"{term}: {postings}")
print(positional_index)
```

```python
N = int(input("Enter the number of queries: "))
for i in range(N):
    query = input("Enter the query sentence: ")
    query = preprocess(query)
    query_words = query.split()
    doc_ids = None

    current_word_positions1 = positional_index[query_words[0]]
    doc_ids = set(current_word_positions1.keys())

    for j in range(1, len(query_words)):
        current_word_positions2 = positional_index[query_words[j]]
        doc_ids2 = set()

        for doc_id in current_word_positions1.keys():
            if doc_id in current_word_positions2:
                positions1 = current_word_positions1[doc_id]
                positions2 = current_word_positions2[doc_id]
                for position1 in positions1:
                    if position1 + 1 in positions2:
                        doc_ids2.add(doc_id)
                        break
        doc_ids = doc_ids2
        current_word_positions1 = current_word_positions2

    doc_ids = list(doc_ids)
    doc_ids = [str(i) + ".txt" for i in doc_ids]
    print(f"Number of documents retrieved for query {i+1} using positional index:", len(doc_ids))
    print(f"Names of documents retrieved for query {i+1} using positional index:", doc_ids)
```
